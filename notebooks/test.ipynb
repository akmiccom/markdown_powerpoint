```python
def convert_marp_to_pandoc(marp_markdown):
    # YAMLメタデータブロックを取得
    yaml_metadata = ""
    if marp_markdown.startswith("---"):
        end_index = marp_markdown.find("---", 3)
        if end_index != -1:
            yaml_metadata = marp_markdown[3:end_index].strip()

    # Markdown本文を取得
    markdown_body = marp_markdown[end_index + 3:].strip()

    # YAMLメタデータをPandocの形式に変換
    pandoc_metadata = ""
    for line in yaml_metadata.split("\n"):
        key, value = line.split(":")
        pandoc_metadata += f"- {key.strip()}: {value.strip()}\n"

    # Markdown本文をPandocの形式に変換
    pandoc_markdown = f"{pandoc_metadata}\n{markdown_body}"

    return pandoc_markdown

# Marp形式のMarkdown
marp_markdown = """
---
marp: true
theme: YPP_ver1.0
paginate: true
# header: motherson yachiyo
# footer: Makoto.Yaguchi@motherson.com_
---

<!-- _class: title -->
# h1 Title Slide タイトルスライド

Makoto.Yaguchi@motherson.com_ 2024/4/20

---

<!-- _class: chapter -->
## h2 Chapter

ここでは○○について解説する

---

<!-- _class: slide -->
### h3 Table of contents
contents
1. Title Slide
2. Title and Content
3. Content with Caption
4. Blank
5. Section Header
6. End page
7. Level 1
8. Level 1
9. Level 1
10. Level 1
11. Level 1
12. Level 1
13. Level 1
14. Level 1

---

<!-- _class: slide -->
### h3 White slide
contents
- Level 1

contents
- Level 2

contents
- Level 3
- Level 1
- Level 1
- Level 1
- Level 1
- Level 1
- Level 1
- Level 1
- Level 1
- Level 1
- Level 1
- Level 1

---

<!-- _class: end -->

© Motherson  All rights reserved by Motherson and/or its affiliated companies. Any commercial use hereof, especially any transfer and/or copying hereof, is prohibited without the prior written consent of Motherson and/or its affiliated companies. In case of transfer of information containing know-how for which copyright or any other intellectual property right protection may be afforded, Motherson and/or its affiliated companies reserve all rights to any such grant of copyright protection and/or grant of intellectual property right protection. www.motherson.com
"""

# Marp形式のMarkdownをPandoc形式に変換
pandoc_markdown = convert_marp_to_pandoc(marp_markdown)
print(pandoc_markdown)
```
